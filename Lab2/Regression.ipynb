```python
import pandas as pd
data_frame = pd.read_csv('LAB2.csv')
X_data_frame = data_frame.iloc[:,:-1]
Y_data_frame = data_frame.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data_frame, Y_data_frame, train_size=0.80, random_state=50)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(X_train, y_train)
y = model.predict(X_test)
y
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure
figure(figsize=(20, 10), dpi=80)

X_graph = np.arange(len(y_test))

plt.scatter(X_graph, y_test)
plt.scatter(X_graph, y)
ax = plt.axes()
plt.ylabel("Y")
plt.xlabel("X")
plt.show()
```
